```python
import math

def sigmoid_func(x):
    return 1/(1+math.exp(-x))
    
```

```python
sigmoid_func(-23)
```

```python
def tanh_func(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh_func(-50)
```

```python
def relu_func(x):
    return max(0,x);
```

```python
relu_func(-7)
```

```python
def leakly_relu_func(x):
    return max(0.1*x,x);
```

```python
leakly_relu_func(-7)
```
